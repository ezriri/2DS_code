```python
# ~ ~ ~ ~ ~ ~ ~ functions ~ ~ ~ ~ ~ ~ ~ 
# for getting useable data from csv files - already decided these thresholds
def reduce_df(big_df):
    smaller_df = big_df[
    (big_df['Euclidean_d_max'] >= 300) &
    (big_df['first_diode_trunc'] == 0) &
    (big_df['last_diode_trunc'] == 0) &
    (big_df['image_trunc'] == 0) &
    (~big_df.isnull().any(axis=1))]  # Exclude rows with NaN
    return smaller_df


# consistency between what I keep extracting - none of these contains name variable, but identifiable by number
def envi_var_1(a_df):
    var_1_df = a_df[['pres']]
    return var_1_df

def envi_var_2(a_df):
    var_2_df = a_df[['pres','wind_w']]
    return var_2_df

def envi_var_3(a_df):
    var_3_df = a_df[['pres','wind_w','LWC']]
    return var_3_df

def envi_var_4(a_df):
    var_4_df = a_df[['pres','wind_w','LWC','IWC']]
    return var_4_df
```

### where are csv files?

```python
data_path = '/gws/nopw/j04/dcmex/users/ezriab/processed_stats/'

ds_csv = '2ds/all_2ds.csv'
hvps_csv = 'hvps/particle_stats/all_hvps.csv'
all_csv = 'all_2ds_hvps.csv'

data_loc = [ds_csv,hvps_csv,all_csv]
key_names = ['ch01','hvps','all']
#sub_key_names = ['df', 'reduced', 'var1','var2','var3','var4']

## loop to make dic of csv file + reduced version + envi var versions too
data_dic = {}
for d in range(3):
    data_dic[key_names[d]]={} 
    loop_df = pd.read_csv(data_path+data_loc[d])
    df_reduced = reduce_df(loop_df)
    
    data_dic[key_names[d]]['df'] = loop_df
    data_dic[key_names[d]]['reduced'] = df_reduced
    data_dic[key_names[d]]['va1'] = envi_var_1(df_reduced)
    data_dic[key_names[d]]['va2'] = envi_var_2(df_reduced)
    data_dic[key_names[d]]['va3'] = envi_var_3(df_reduced)
    data_dic[key_names[d]]['va4'] = envi_var_4(df_reduced)
```

```python
## e.g. 
#data_dic['ch01']['reduced']
```

## scale the data!

```python
## make another loop for scaling data
var_names = ['va1','va2','va3','va4']
scale_func_lst = [StandardScaler(), MinMaxScaler(),RobustScaler()]
scale_dic_names = ['Standard', 'MinMax', 'Robust']

scaled_dic = {}
for d in range(3):
    scaled_dic[key_names[d]]={}
    for var in var_names:
        var_data = data_dic[key_names[d]][var]
        scaled_dic[key_names[d]][var] = {}
        for i in range(len(scale_func_lst)):
            scale_function = scale_func_lst[i]
            scaled_var_data = scale_function.fit_transform(var_data)
            scaled_df = pd.DataFrame(scaled_var_data, columns=var_data.columns)
            scaled_dic[key_names[d]][var][scale_dic_names[i]] = scaled_df
```

```python
save_loc = '/home/users/esree/2DS_code/unsupervised_ML/plots/'

key_names = ['ch01','hvps','all']
#envi_var_names = ['va1','va2','va3','va4']
envi_var_names = ['va2','va3','va4']

scale_dic_names = ['Standard', 'MinMax', 'Robust']
scale_plt_names = ['SS', 'MMS', 'RS']

for data_n in key_names:
    for var_n in envi_var_names:
        for i in range(3):
            data_df = scaled_dic[data_n][var_n][scale_dic_names[i]]
            save_name = f'{data_n}_{var_n}_PCA2_{scale_plt_names[i]}.png'
            title_n = f'First two PCA dimensions of {data_n} {var_n} {scale_plt_names[i]}'
            # PCA time
            x_pca_2 = PCA(n_components=2).fit_transform(data_df)
            
            # make plot
            fig, ax = plt.subplots(figsize=(8, 6))
            
            scatter = ax.scatter(
                x_pca_2[:, 0],
                x_pca_2[:, 1],
            )
            
            ax.set(
                title=title_n,
                xlabel="1st Eigenvector",
                ylabel="2nd Eigenvector",
            )
            ax.xaxis.set_ticklabels([])
            ax.yaxis.set_ticklabels([])
            
            fig.savefig(save_loc+save_name)
            plt.close(fig)
```
